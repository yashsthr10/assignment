Identifying the most profitable products

```python
data = pd.read_csv('transaction_data.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data.head()
```

```python
# TOP PRODUCTS
top_products = data.groupby("PROD_NAME")["TOT_SALES"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind="bar", stacked=True, figsize=(16,8))
plt.title("TOP PRODUCTS")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.legend(title="Premium Category")
plt.savefig("top-products.png")
plt.show()
```

```python
#TOP CUSTOMERS
data.groupby("LYLTY_CARD_NBR")["TOT_SALES"].sum().sort_values(ascending=False).head(10)
```

```python
#TOP MOST FREQUENT CUSTOMERS
frequent_customers = data.groupby("LYLTY_CARD_NBR")["TXN_ID"].nunique().sort_values(ascending=False)
frequent_customers.head(10)
```

```python
#
loyalty_df = data.groupby("LYLTY_CARD_NBR").agg(
    total_spending=("TOT_SALES", "sum"),
    total_transactions=("TXN_ID", "nunique"),
    unique_products=("PROD_NBR", "nunique"),
    total_quantity=("PROD_QTY", "sum")
)

loyalty_df = (loyalty_df - loyalty_df.min()) / (loyalty_df.max() - loyalty_df.min())

loyalty_df["loyalty_score"] = (
    0.4 * loyalty_df["total_transactions"] + 
    0.3 * loyalty_df["total_spending"] + 
    0.2 * loyalty_df["unique_products"] + 
    0.1 * loyalty_df["total_quantity"]
)

loyalty_df = loyalty_df.sort_values("loyalty_score", ascending=False)
print(loyalty_df.head(10)) 
```

```python
plt.figure(figsize=(10, 5))
loyalty_df["loyalty_score"].head(10).plot(kind="bar")
plt.title("Top 10 Most Loyal Customers")
plt.xlabel("LYLTY_CARD_NBR")
plt.ylabel("Loyalty Score")
plt.savefig("loyalCustomers.png")
plt.show()

```

```python
# Merge the two datasets on LYLTY_CARD_NBR
customer_loyalty = data.groupby("LYLTY_CARD_NBR").agg(
    total_spending=("TOT_SALES", "sum"),
    total_transactions=("TXN_ID", "nunique"),
    unique_products=("PROD_NBR", "nunique"),
    total_quantity=("PROD_QTY", "sum")
).reset_index()

# Merge with customer demographics dataset (containing LIFESTAGE and PREMIUM_CUSTOMER)
merged_df = customer_loyalty.merge(data1, on="LYLTY_CARD_NBR", how="left")
```

```python
lifestage_loyalty = merged_df.groupby("LIFESTAGE").agg(
    avg_spending=("total_spending", "mean"),
    avg_transactions=("total_transactions", "mean"),
    avg_unique_products=("unique_products", "mean"),
    avg_quantity=("total_quantity", "mean")
).reset_index()
```

```python


plt.figure(figsize=(10, 5))
sns.barplot(data=lifestage_loyalty, x="LIFESTAGE", y="avg_spending", palette="viridis")
plt.xticks(rotation=15)
plt.title("Average Spending by Lifestage")
plt.xlabel("Lifestage")
plt.ylabel("Average Spending")
plt.savefig("LSvsAvgSpndg.png")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=lifestage_loyalty, x="LIFESTAGE", y="avg_transactions", marker="o")
plt.xticks(rotation=45)
plt.title("Average Transactions per Lifestage")
plt.xlabel("Lifestage")
plt.ylabel("Average Transactions")
plt.show()
```
